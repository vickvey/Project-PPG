```python

# Define the mapping dictionary
mapping = {1: 5, 2: 4, 3: 3, 4: 2, 5: 1, None: None}
```

```python
# Apply the mapping using map() function
df['AS1_2_changed'] = df['AS1_2'].map(mapping)
```

```python
test_unique_or_not = df['AS1_2'] + df['AS1_2_changed']
```

```python
df['AS1_2'] = df['AS1_2_changed']
```

```python
df.drop(columns=['AS1_2_changed'], inplace=True)
```

```python
# Select only the columns you want to sum
columns_to_sum = ['AS1_1', 'AS1_2', 'AS1_3', 'AS1_4', 'AS1_5']
```

```python
# Sum values across rows for selected columns and store the result in a new column `anxiety_meter`
df['anxiety_meter'] = df[columns_to_sum].sum(axis=1)
```

```python
# Create 'Anxious' column based on sum condition
df['Anxious'] = (df['anxiety_meter'] >= 16).astype(int)
```

```python
# List of IDs to remove
missing = [110, 131, 136, 174, 182, 185, 197, 206, 177]
# Remove rows where 'P_Id' is in the 'missing' list
df = df[~df['P_Id'].isin(missing)]
```

```python
# Create a new DataFrame with 'P_Id', 'Anxious' and 'anxiety_meter' columns
result_df = df[['P_Id', 'anxiety_meter', 'Anxious']]
```
